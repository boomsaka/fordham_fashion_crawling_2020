```python
from bs4 import BeautifulSoup
import urllib.request
from datetime import date
import pandas as pd
date_today = date.today()
```

```python
product_url = []
name = []
sales_price = []
price = []
picture_url = []
brand = []
product_type = []
timestamp  = []

#items = ['women-dresses', 'women-outerwear', 'women-jackets', 'women-jeans', 
#         'women-pants', 'women-skirts-shorts', 'women-t-shirts', 'women-blouses-shirts',
#        'women-knitwear', 'women-sweatshirts', 'women-jumpsuits']

items = ['men-outerwear', 'men-jackets', 'men-jeans', 
         'men-pants', 'men-suits', 'men-shorts', 'men-t-shirts', 'men-shirts',
        'men-knitwear', 'men-sweatshirts']
for category in items:
    url = 'https://www.apc-us.com/collections/' + category
    print(url)

    ourUrl = urllib.request.urlopen(url)
    soup = BeautifulSoup(ourUrl,'html.parser')
    for i in soup.find_all('div',{'class':'plp-container product-grid margin-top--xs col-6 col-4-md col-4-lg'}):
        product_url.append('http://apc-us.com' + i.find('a',{'class':'item plp-container__main-url'}).get('href'))
        name.append(i.find('div',{'class':'caption-default'}).find('div',{'class':"name"}).text)
        if i.find('div',{'class':'caption-default'}).find('div',{'class':"price"}).find('span',{'class':'color--red'}) != None:
            sales_price.append(str(i.find('div',{'class':'caption-default'}).find('div',{'class':"price"}).find('span',{'class':'color--red'}).text).strip())
        else:
            sales_price.append(str(i.find('div',{'class':'caption-default'}).find('div',{'class':"price"}).text).strip())
        
        if i.find('div',{'class':'caption-default'}).find('div',{'class':"price"}).find('span',{'class':'line-through margin-right--sm'}) != None:
            price.append(i.find('div',{'class':'caption-default'}).find('div',{'class':"price"}).find('span',{'class':'line-through margin-right--sm'}).text)
        else:
            price.append(None)
        
        picture_url.append('https:' + i.find('div',{'class':'reveal image'}).img.get('src'))
        brand.append('APC')
        product_type.append(category.replace('-', ' '))
        timestamp.append(date_today)
    
```

```python
column_names = ["brand", "product_type", "product_name","price", 
                "sales_price", "product_url", "picture_url", "timestamp"]
first_pass = pd.DataFrame(columns = column_names)
first_pass.brand = brand
first_pass.product_type = product_type
first_pass.product_name = name
first_pass.price = price
first_pass.sales_price = sales_price
first_pass.product_url = product_url
first_pass.picture_url = picture_url
first_pass.timestamp = timestamp
first_pass
```

```python
# Back up
#first_pass.to_csv('apc-women.csv', encoding='utf-8-sig')
first_pass.to_csv('apc-men.csv', encoding='utf-8-sig')
```

# Second Layer: Description, Color, Material

```python
second_layer = pd.read_csv("apc-men.csv", encoding = 'utf-8-sig').drop(['Unnamed: 0'], axis=1)
```

```python
color = []
product_description = []
material = []
for item in second_layer.product_url.to_list():
    ourUrl = urllib.request.urlopen(item)
    soup_detail = BeautifulSoup(ourUrl,'html.parser')
    color.append(soup_detail.find('li',{'class':'specifics-color'}).find('span',{'id':'selectVariantColorDisplayName'}).text)
    product_description.append(str(soup_detail.find('div',{'class':'description'}).text).strip())
    material.append(None)
```

```python
second_layer.insert(3, "product_description", product_description, True) 
second_layer.insert(6, "color", color, True) 
second_layer.insert(7, "material", material, True)
```

```python
second_layer.to_csv('apc-men-detail.csv', encoding='utf-8-sig')
```
