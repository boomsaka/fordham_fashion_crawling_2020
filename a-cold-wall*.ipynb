```python
from bs4 import BeautifulSoup
import urllib.request
from datetime import date
import pandas as pd
date_today = date.today()
```

```python
product_url = []
name = []
sales_price = []
price = []
picture_url = []
brand = []
product_type = []
timestamp  = []

url = 'https://a-cold-wall.com/collections/all'
ourUrl = urllib.request.urlopen(url)
soup = BeautifulSoup(ourUrl,'html.parser')
for i in soup.find_all('div',{'class':'product-card'}):
    
    timestamp.append(date_today)
    price.append(None)
    brand.append('A Cold Wall*')
    product_type.append(None)
    
    name.append(str(i.find('figcaption',{'class':'mt-4 text-center uppercase leading-relaxed'}).span.text).strip())
    sales_price.append(str(i.find('figcaption',{'class':'mt-4 text-center uppercase leading-relaxed'}).div.text).strip())
    product_url.append("https://a-cold-wall.com" + i.find('a').get('href'))
    picture_url.append("https:"+str(i.find('figure',{'class':'select-none aspect-ratio-5/7'}).div.img.get('data-src')).replace('{width}','600'))
    
```

# Second Layer

```python
color = []
product_description = []
material = []
picture_name = []
counter = 1

for item in product_url:
    
    ourUrl = urllib.request.urlopen(item)
    soup = BeautifulSoup(ourUrl,'html.parser')
    picture_name.append('a-cold-wall_' + str(counter) + '.jpg')
    product_description.append(soup.find('div',{'class':'product-description rte mb-4'}).text.replace('Show details','').replace('Hide details','').strip().replace('\n',''))
    color.append(None)
    material.append(None)
    counter += 1
    
```

```python
column_names = ["brand", "product_type", "product_name", 'product_description', "price", 
                "sales_price", "color","material", "product_url", "picture_url", "timestamp", 'picture_name']
second_pass = pd.DataFrame(columns = column_names)
second_pass.brand = brand
second_pass.product_type = product_type
second_pass.product_name = name
second_pass.product_description = product_description
second_pass.price = price
second_pass.sales_price = sales_price
second_pass.color = color
second_pass.material = material
second_pass.product_url = product_url
second_pass.picture_url = picture_url
second_pass.timestamp = timestamp
second_pass.picture_name = picture_name
second_pass.head()
```

```python
second_pass.to_csv('a-cold-wall.csv', index = False, encoding = 'utf-8-sig')
```
