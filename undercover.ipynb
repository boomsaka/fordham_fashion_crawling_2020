```python
from bs4 import BeautifulSoup
import urllib.request
from datetime import date
import pandas as pd
date_today = date.today()

men_items = ['tops', 'outer', 'tshirt', 'bottoms']
men_items_pages = [5, 3, 2, 2]
women_items_pages = [2, 1, 1, 1]
women_items = ['tops', 'outer', 'tshirt', 'bottoms']

name = []
product_url = []
picture_url = []
brand = []
timestamp = []
product_type = []
price = []
sales_price = []

for index in range(len(women_items)):
    counter = 1
    for p in range(women_items_pages[index]):
        if counter <= women_items_pages[index]:
            ourUrl = urllib.request.urlopen('https://store.undercoverism.com/undercoverwomen/' 
                                            + women_items[index] + '?order_by=1&is_secret=1&page=' + str(counter))
            counter += 1
            soup = BeautifulSoup(ourUrl,'html.parser')
            
            for i in soup.find('div',{'class':'itemList'}).find_all('li'):
                brand.append('Undercover')
                timestamp.append(date_today)
                product_type.append("Women's Clothing")
                
                name.append(i.find('div',{'class':'commodityNumberWrapper'}).text)
                product_url.append("https://store.undercoverism.com" + i.find('div',{'class':'itemImage goods_overlap_image'}).a.get('href'))
                picture_url.append("https://store.undercoverism.com" + i.find('div',{'class':'itemImage goods_overlap_image'}).img.get('src'))
                
                price.append(None)
                sales_price.append(i.find('div',{'class':'priceWrapper clearfix'}).span.text)
```

# Second Layer

```python
material = []
product_description = []
picture_name = []
color = []
counter = 1

for link in product_url:

    ourUrl = urllib.request.urlopen(link)
    soup = BeautifulSoup(ourUrl,'html.parser')
    
    picture_name.append('undercover-women_' + str(counter) + '.jpg') # change variable
    counter += 1
    
    product_description.append(None)
    color.append(None)
    material.append(soup.find('div',{'class':'material'}).dd.text.strip())
    
```

```python
column_names = ["brand", "product_type", "product_name", 'product_description', "price", 
                "sales_price", "color","material", "product_url", "picture_url", "timestamp", "picture_name"]
second_pass = pd.DataFrame(columns = column_names)
second_pass.brand = brand
second_pass.product_type = product_type
second_pass.product_name = name
second_pass.product_description = product_description
second_pass.price = price
second_pass.sales_price = sales_price
second_pass.color = color
second_pass.material = material
second_pass.product_url = product_url
second_pass.picture_url = picture_url
second_pass.timestamp = timestamp
second_pass.picture_name = picture_name
second_pass.head()
```

```python
second_pass.to_csv("undercover-women.csv", index=False, encoding = 'utf-8-sig')
```
