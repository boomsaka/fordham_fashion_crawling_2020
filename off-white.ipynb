```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time

from datetime import date
import pandas as pd
date_today = date.today()

product_url = []
picture_url = []
name = []
sales_price = []
price = []
brand = []
product_type = []
timestamp = []

options = webdriver.ChromeOptions()
options.add_argument("start-maximized")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument('--disable-blink-features=AutomationControlled')

driver = webdriver.Chrome(options=options, executable_path=r'/Users/mandili/.wdm/drivers/chromedriver/87.0.4280.20/mac64/chromedriver')
for page in range(1, 6): # men:8

    #driver.get("https://www.off---white.com/en-us/shopping/man/clothing?pageIndex="+str(page))
    driver.get('https://www.off---white.com/en-us/shopping/woman/clothing?pageIndex='+str(page))
    time.sleep(3)
    soup = BeautifulSoup(driver.execute_script("return document.body.innerHTML;"),'html.parser')
    #driver.close()
    
    for i in soup.find_all('li',{'class':'css-78is7c e1qfmkbd1'}):
    
        brand.append('Off White')
        product_type.append("Women's Clothing") # change variable
        timestamp.append(date_today)
        product_url.append("https://www.off---white.com" + i.find('a',{'data-test':'productCard-container'}).get('href'))
        picture_url.append(i.find('a',{'data-test':'productCard-container'}).img.get('src'))
        name.append(i.find('p',{'data-test':'productCard-productName'}).text.lower())
    
        if i.find('div',{'itemprop':'price'}).find('span',{'data-test':'product-salePrice'}) != None:
            sales_price.append(i.find('div',{'itemprop':'price'}).find('span',{'data-test':'product-salePrice'}).text)
            price.append(i.find('div',{'itemprop':'price'}).find('span',{'aria-label':'regular price'}).text)
        else:
            price.append(None)
            sales_price.append(i.find('div',{'itemprop':'price'}).find('span',{'aria-label':'regular price'}).text)
```
