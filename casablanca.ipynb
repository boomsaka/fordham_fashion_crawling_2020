```python
from bs4 import BeautifulSoup
import urllib.request

from datetime import date
import pandas as pd
date_today = date.today()

name = []
product_url = []
sales_price = []
price = []
timestamp  = []
brand = []
product_type = []
image_url = []

ourUrl = urllib.request.urlopen('https://casablancaparis.com/collections/all')
soup = BeautifulSoup(ourUrl,'html.parser')

for i in soup.find_all('a',{'class':'product-list-item'}):
    price.append(None)
    timestamp.append(date_today)
    brand.append('casablanca')
    product_type.append('')
    name.append(i.find('div',{'class':'product-title-container'}).h3.text)
    index_price = str(i.find('div',{'class':'product-title-container'})).find('</h3>')
    sales_price.append(str(i.find('div',{'class':'product-title-container'}))[index_price+5:].replace('</div>','').strip())
    product_url.append('https://casablancaparis.com' + i.get('href'))
    image_url.append("https:" + str(i.find('div',{'class':'product-image'}).img.get('src')).replace('300x.', '600x.'))
```

```python
column_names = ["brand", "product_type", "product_name", "price", 
                "sales_price", "product_url", "picture_url", "timestamp"]
first = pd.DataFrame(columns = column_names)
first.brand = brand
first.product_type = product_type
first.product_name = name
#first.product_description = product_description
first.price = price
first.sales_price = sales_price
#first.color = color
#first.material = material
first.product_url = product_url
first.picture_url = image_url
first.timestamp = timestamp
first.head()
```

# Second Pass

```python
color = []
product_description = []
material = []
picture_name = []
counter = 1

for link in product_url:
    ourUrl = urllib.request.urlopen(link)
    soup = BeautifulSoup(ourUrl,'html.parser')
    product_description.append(str(soup.find('div',{'class':'product-info-description'}).text).strip())
    material.append(None)
    picture_name.append(brand[0]+'_'+str(counter)+'.jpg')
    color.append(None)
    counter += 1
```

```python
column_names = ["brand", "product_type", "product_name", 'product_description', "price", 
                "sales_price", "color","material", "product_url", "picture_url", "timestamp", 'picture_name']
second_pass = pd.DataFrame(columns = column_names)
second_pass.brand = brand
second_pass.product_type = product_type
second_pass.product_name = name
second_pass.product_description = product_description
second_pass.price = price
second_pass.sales_price = sales_price
second_pass.color = color
second_pass.material = material
second_pass.product_url = product_url
second_pass.picture_url = image_url
second_pass.timestamp = timestamp
second_pass.picture_name = picture_name
second_pass.head()
```

```python
second_pass.to_csv('casablanca.csv', index = False, encoding = 'utf-8-sig')
```
