```python
import requests
import re
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
SCROLL_PAUSE_TIME = 1.2


URL = 'https://www.underarmour.com/en-us/d/mens/clothing/'
driver = webdriver.Chrome(ChromeDriverManager().install())


driver.get(URL)

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

soup = BeautifulSoup(driver.page_source,'html.parser')
```

```python
with open("11.htm", 'r') as f:

    contents = f.read()

    soup = BeautifulSoup(contents, 'lxml')
product_name = []
sale_price = []
original_price = []
product_url = []
image_url = []
for i in soup.find_all('section',{'class':'b-products_grid-tile'}):
    product_name.append(i.find('a',{'class':'b-tile-name'}).text)
    if i.find('div',{'class':'b-price'}).find('span',{'class':'b-price-value m-strikethrough bfx-price highlighted'}) != None:
        original_price.append(str(i.find('div',{'class':'b-price'}).find('span',{'class':'b-price-value m-strikethrough bfx-price highlighted'}).text).strip())
    else:
        original_price.append(None)
    if (i.find('div',{'class':'b-price'}).find('span',{'class':'b-price-value highlighted bfx-price m-actual'})) != None:
        sale_price.append(str(i.find('div',{'class':'b-price'}).find('span',{'class':'b-price-value highlighted bfx-price m-actual'}).text).strip())
    elif i.find('div',{'class':'b-price'}).find('span',{'class':'b-price-value bfx-price'}) != None:
        sale_price.append(str(i.find('div',{'class':'b-price'}).find('span',{'class':'b-price-value bfx-price'}).text).strip())
    else:
        sale_price.append(None)
    product_url.append('https://www.underarmour.com'+str(i.find('a',{'class':'b-tile-name'}).get('href')))
    if (i.find('a',{'class':'b-tile-images_container'}).find('img',{'class':'b-tile-image b-tile-main_image js-tile-carousel_image'})) !=None:
        image_url.append(i.find('a',{'class':'b-tile-images_container'}).find('img',{'class':'b-tile-image b-tile-main_image js-tile-carousel_image'}).get('src'))
    else:
        image_url.append(None)
```

```python
import pandas as pd

column_names = ["product_name", "original_price", "sale_price",'product_link', 'image_link']
df = pd.DataFrame(columns = column_names)


df['product_name'] = product_name
df['original_price'] = original_price
df['sale_price'] = sale_price
df['image_link'] = image_url
df['product_link'] = product_url
```

```python
df.to_csv('under-armour-men.csv', encoding='utf-8-sig')
```

```python
import pandas as pd
import requests
import re
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
detail_df = pd.read_csv('under-armour-women.csv', encoding= 'unicode_escape')
description = []
color = []
material = []

for i in range(349, len(detail_df.product_link)):
    URL = detail_df.product_link.to_list()[i]
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(URL)
    
    color_i = []
    des_i = []
    mat_i = []
# create a BeautifulSoup object which represents the document as a nested data structure
    soup = BeautifulSoup(driver.page_source,'html.parser')
    #driver.quit()
    
    i = soup.find('ul',{'class':'b-swatches_circle swiper-wrapper js-swiper-wrapper'})
    if i != None:
        for i in i.find_all('li'):
            color_i.append(i.find('a').get('title'))
        color.append(color_i)
    else:
        color.append(None)
    
    
    details_i = soup.find('div',{'class':'b-product_description-items g-accordion--mobile g-tabs-content'})
    if details_i != None:
        for i in details_i.find('div',{'id':'tab-1'}).find_all('li'):
            des_i.append(str(i.text).strip().replace("/n",''))
        description.append(des_i)
    else:
        description.append(None)
        
    material_i = soup.find('div',{'class':'b-product_description-items g-accordion--mobile g-tabs-content'})
    if material_i != None:
        for i in material_i.find('div',{'id':'tab-2'}).find_all('li'):
            mat_i.append(str(i.text).strip().replace("/n",''))
        material.append(mat_i)
    else: 
        material.append(None)
```

```python
import pandas as pd

column_names = ["color", "description", "material"]
df = pd.DataFrame(columns = column_names)

df['color'] = color
df['description'] = description
df['material'] = material
```

```python
df.to_csv('details-under-amour-women.csv', encoding='utf-8-sig')
```
